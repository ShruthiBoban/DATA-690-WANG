The column 'Census Year' has bad data as astericks present in them.

The column 'Total Population' looks good.

The column 'Census Cost' has '$' symbols which needs to be removed along with string present in the last few rows which needs to be removed and zeroes to be added in place of them to convert the column into int type.

The column 'Average Cost per Person' has 'cents' and '$' symbols present in the rows which needs to be removed.

## Write a function named cleanse_year() that takes a string as input and removes any asterisks in the string and return the cleansed string. Test the function by using test strings (for example, "1989*", "20*10")

## Cleanse the column "Census Year" using the function cleanse_year() and Pandas's apply() function.

```
def cleanse_year(census_year):
    year = census_year.replace('*','')
    return year
```

```
df['Year'] = df['Census Year'].apply(cleanse_year)
df
```

## Repeat this process for column "Total Population", "Census Cost", "Average Cost Per Person". (define, test, and apply the function)

```
df['Population'] = df['Total Population']
df
```

```
def cleanse_cost(census_cost):
    cost = census_cost.lstrip("$").replace(",","")
    if  census_cost.find("4.5") != -1:          # if "4.5" in raw_cost:
        cost = "45" + "0" * 8
    elif census_cost.find("Billion") != -1:     # elif "Billion" in raw_cost: 
        cost = "13" + "0" * 9
    return cost
```

```
df['Cost'] = df['Census Cost'].apply(cleanse_cost)
df
```

```
def cleanse_avgcost(avg_cost_person):
    avg = avg_cost_person.lstrip("$").replace("cents","")
    return avg
```

```
df['Average Cost'] = df['Average Cost Per Person'].apply(cleanse_avgcost)
df
```

```
df.dtypes
```

```
df["Year"] = df["Year"].astype(int)
df["Population"] = df["Population"].astype(int)
df["Cost"] = df["Cost"].astype(int)
df["Average Cost"] = df["Average Cost"].astype(float)
```

```
df.dtypes
```

```
df = df.drop(['Census Year','Total Population','Census Cost','Average Cost Per Person'], axis = 1)
```
